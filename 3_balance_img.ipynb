```python
import os
import math
import random
import cv2 as cv
import common as c
```

```python
img_list = c.get_r3_img_list()

for lb in c.IMG_LABEL:
    path = os.path.join(c.RZD_3_PATH, lb)
    print(f'{lb.replace("_", " ").title():<10}', ':', len(c.get_imgs([], path)))
print("Total IMG  :", len(img_list))
print()
DIF = len(img_list) - 2*len(c.get_imgs([], c.RZD_3_HEALTHY))
print(f'{"DIFF":<10}',  ':', DIF)
```

```python
extra_img_list = os.listdir(c.RD_PATH)
if len(extra_img_list) - DIF < 0:
    print("ERROR: Too much bias in data.")
new_img = random.choices(extra_img_list, k=DIF)
print("NEW IMG : ",len(new_img))
```

```python
def get_type_from_path(img_n, count):
    return os.path.join(c.RZD_3_HEALTHY, c.IMG_LABEL[c.C_HEALTHY] + '_' + str(count) + ".jpg")

def resize_img(img_c, batch_list):
    count = img_c
    for i,img_n in enumerate(batch_list):
        img = os.path.join(c.RD_PATH, img_n)
        print("Processing Batch: ", count, i, len(batch_list))
        #break;
        mat_img = cv.imread(img, cv.IMREAD_COLOR)
        try:
            height, width = mat_img.shape[:2]
            #print("ORG : ", height, width)
            if width < height:
                factor = c.IMG_SIZE/width
                n_width = c.IMG_SIZE
                n_height = round(height * factor)
                p_img_count = math.ceil(n_height/c.IMG_SIZE)
                rez_img = cv.resize(mat_img, (n_width, n_height), interpolation = cv.INTER_AREA)
                for i in range(p_img_count):
                    if n_height < c.IMG_SIZE:
                        n_height = c.IMG_SIZE
                    end = n_height
                    start = n_height - c.IMG_SIZE
                    n_height -= c.IMG_SIZE
                    n_mat = rez_img[start:end, :, :]
                    cv.imwrite(get_type_from_path(img_n, count), n_mat)
                    count += 1
            else:
                factor = c.IMG_SIZE/height
                n_width = round(width * factor)
                n_height = c.IMG_SIZE
                p_img_count = math.ceil(n_width/c.IMG_SIZE)
                rez_img = cv.resize(mat_img, (n_width, n_height), interpolation = cv.INTER_AREA)
                for i in range(p_img_count):
                    if n_width < c.IMG_SIZE:
                        n_width = c.IMG_SIZE
                    end = n_width
                    start = n_width - c.IMG_SIZE
                    n_width -= c.IMG_SIZE
                    n_mat = rez_img[:, start:end, :]
                    cv.imwrite(get_type_from_path(img_n, count), n_mat)
                    count += 1
        except Exception as e:
            print("Error -", e)
            cv.imwrite(os.path.join(c.DUMP_PATH, img), mat_img)

healthy_count = len(c.get_imgs([], c.RZD_3_HEALTHY))
resize_img(healthy_count, new_img)
```
